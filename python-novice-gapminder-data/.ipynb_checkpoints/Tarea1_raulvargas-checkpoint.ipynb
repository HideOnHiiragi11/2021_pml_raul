## Variables and Assignment: Swapping Values

```python
# Command  # Value of x   # Value of y   # Value of swap #
x = 1.0    # 1.0          # -            # -             #
y = 3.0    # 1.0          # 3.0          # -             #
swap = x   # 1.0          # 3.0          # 1.0           #
x = y      # 3.0          # 3.0          # 1.0           #
y = swap   # 3.0          # 1.0          # 1.0           #
```

## For Loops: Cumulative Sum

```python
data = [1,2,2,5]
total = 0
cumulative = []
for number in data:
    total += number
    cumulative.append(total)
print(cumulative)
```

## Conditionals: Trimming Values

```python
original = [-1.5, 0.2, 0.4, 0.0, -1.3, 0.4]
result = []
for value in original:
    if value < 0.0:
        result.append(0)
    else:
        result.append(1)
print(result)
```
